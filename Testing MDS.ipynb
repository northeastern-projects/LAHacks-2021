```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import MDS
```

```python
test_data = np.random.normal(loc = 10, scale = 3, size=(100, 10))
```

```python
test_mds = MDS(n_components=3)
```

```python
D3 = test_mds.fit_transform(test_data)
```

```python
D3
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
xs = D3[:,0]
ys = D3[:,1]
zs = D3[:,2]
ax.scatter(xs, ys, zs)
plt.show()
```

```python
print(np.mean(D3), np.std(D3))
```

```python
print(np.mean(test_data), np.std(test_data))
```
